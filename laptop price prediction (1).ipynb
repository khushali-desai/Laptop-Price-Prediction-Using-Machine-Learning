```python
num_columns = len(df.select_dtypes(include='number').columns)

fig, axes = plt.subplots(1, num_columns, figsize=(35, 10))

for i, column in enumerate(df.select_dtypes(include='number').columns):
    sns.boxplot(data=df[column], ax=axes[i])
    axes[i].set_title(f'Boxplot of {column}')
    axes[i].set_ylabel(column)
    
plt.show()
```

```python
#here the maximun outliers are in price column so we can fix it using iqr method 
```

```python
uq,lq=np.percentile(df["Price"],(75,25))
```

```python
uq
```

```python
lq
```

```python
iqr=uq-lq
iqr
```

```python
ub=uq+1.5*iqr
lb=lq-1.5*iqr
print(ub)
print(lb)
```

```python
sns.boxplot(df[(df["Price"]<ub)&(df["Price"]>lb)]["Price"],color="purple")
```

```python
df = df[(df['Price']>=(lq -1.5*iqr)) & (df['Price'] <= (uq +1.5*iqr))]
```

```python
#8.In cases where categorical variables are present, how can they be encoded using appropriate techniques?
```

```python
cetegorycal_column=df.select_dtypes(include=["object"]).columns
for col in cetegorycal_column:
    df[col].fillna(df[col].mode()[0], inplace=True)
    
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    df[col] =enc.fit_transform(df[col])
print(df[categorical_columns].head())
print(df.dtypes)
```

```python
#14. Proper approach to enhance the model by certain techniques must be done?
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5]}

grid_search = GridSearchCV(RandomForestRegressor(), param_grid, cv=5)
grid_search.fit(x_train, y_train)
print(grid_search.best_params_)
```
